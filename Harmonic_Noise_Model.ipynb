# Modelling noise recieved by MEMS microphones

```python
import numpy as np
import matplotlib.pyplot as plt
import IPython
from scipy.io import wavfile
```

```python
#Processing speech signal 
Fs, speech = wavfile.read('Speech_10sec.wav')
desired_sec =10
speechL = speech[:,0]
speechR = speech[:,1]
speechL = speechL[0:desired_sec*Fs]
x = len(speech)

X = np.linspace(1,x,x)

time = x/Fs
time_axis = X/Fs 

T = 1/Fs
L = len(time_axis)

#synthesis harmonic noise signal 
fund_freq = 70 # Fundamental frequency 
t = np.linspace(1,desired_sec,(Fs*desired_sec)) 

num_of_har = 9*2
num_of_per = 1+1
num_of_gauss = 2
aph = 1/num_of_per
harmonics = np.zeros([1,num_of_har])

for a in range(0,num_of_har):
    harmonics[0,a] = fund_freq *(a+1)

sd = 1;
m = round(num_of_har/(2*num_of_gauss));
y = np.linspace(1,(num_of_har/2),(num_of_har/2))

guass = np.zeros([num_of_gauss,len(y)+1])

 
for i in range(0,num_of_gauss):
    for n in range(0,len(y)):
        guass[i,n] = (1/(sd * np.sqrt(2*np.pi))) * np.exp(-((y[n]-m)**2) / (2*(sd)**2))


harmonica = np.zeros([num_of_har,len(t)])
har_per = int(num_of_har/2)
for n in range(0,num_of_gauss):
    for i in range(0,har_per):
        harmonica[(i+(n*har_per)),:] = guass[n,i] * np.sin(2*np.pi*harmonics[0,i+(n*har_per)]*t);


H = (aph*(np.sin(2*np.pi*t*fund_freq)) + 0.6*aph*(sum(harmonica[0:int(num_of_har/2),:])) +  0.4*aph*(sum(harmonica[int(num_of_har/2):-1,:])))*0.05;

#adding speech and noise 
mixing = H + speechL
```

# Results

```python
print("Clean Speech:")
IPython.display.Audio(speechL, rate=Fs)
```

```python
print("Noise:")
IPython.display.Audio(H, rate=Fs)
```

```python
print("Linear mixing:")
IPython.display.Audio(mixing, rate=Fs)
```
